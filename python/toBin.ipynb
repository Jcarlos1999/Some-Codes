```python
def toBin(x):
  aux = ''
  while True:
    div = x%2
    x = int(x / 2)
    if div == 0:
      aux = aux + '0'
    elif div == 1:
      aux = aux + '1'
    if x == 0:
      break
    elif x == 1:
      aux = aux + '1'
      break

  return aux

def toDecimal(x):
  xBin = str(x)
  dec = 0
  aux = len(xBin) - 1
  print(xBin)
  for i in xBin:
    if i == '1':
      dec += 2**aux 
    aux -= 1
  return dec

y = toDecimal(10000)
#print(f'Valor binario de {y[::-1]}')
print(f'Valor do decimal: {y}')
```
